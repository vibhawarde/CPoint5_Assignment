```python
result_df = pd.concat([initial_inventory.groupby('month').agg({'Inventory':'sum'}).reset_index(), inventory_addition.groupby('month').agg({'Total Item Qty':'sum'}).reset_index()], ignore_index=True)
```

```python
result_df['Inventory'] = result_df['Inventory'].fillna(result_df['Total Item Qty'])
```

```python
result_df=result_df.drop(['Total Item Qty'],axis=1)
```

```python
result_df['cross_key'] = 1
c=pd.merge(result_df,result_df,how='inner',on='cross_key')
c[c['month_x']>=c['month_y']]
d=c[c['month_x']>=c['month_y']].groupby('month_x').agg({'Inventory_y':'sum'}).reset_index()
```

```python

plt.figure(figsize=(8, 6))
plt.bar(d['month_x'], d['Inventory_y'], color='skyblue', width=15)

plt.xlabel('Month')
plt.ylabel('Inventory Level')
plt.title('Bar Chart ')
plt.grid(False)


plt.show()
```
